```python

import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torchaudio
from torchaudio.transforms import Resample
import numpy as np
import gc
import random

class SpectralConvergenceLoss(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, pred_mag, target_mag):
        return torch.norm(target_mag - pred_mag, p='fro') / torch.norm(target_mag, p='fro')

class STFTLoss(nn.Module):
    def __init__(self, fft_size=1024, hop_length=256, win_length=1024):
        super().__init__()
        self.fft_size = fft_size
        self.hop_length = hop_length
        self.win_length = win_length
        self.window = torch.hann_window(win_length)
        self.spectral_convergence = SpectralConvergenceLoss()
        self.l1 = nn.L1Loss()

    def stft(self, y):
        stft_out = torch.stft(y, self.fft_size, hop_length=self.hop_length,
                              win_length=self.win_length, window=self.window.to(y.device),
                              return_complex=False)  # shape: (B, F, T, 2)
        real = stft_out[..., 0]
        imag = stft_out[..., 1]
        mag = torch.sqrt(real ** 2 + imag ** 2)
        return mag

    def forward(self, y_hat, y):
        y_hat = y_hat.squeeze(1)
        y = y.squeeze(1)
        y_hat_mag = self.stft(y_hat)
        y_mag = self.stft(y)
        sc_loss = self.spectral_convergence(y_hat_mag, y_mag)
        mag_l1_loss = self.l1(y_hat_mag, y_mag)
        return sc_loss + mag_l1_loss

class ResBlock(nn.Module):
    def __init__(self, channels, kernel_size=3, dilation=1):
        super().__init__()
        self.block = nn.Sequential(
            nn.LeakyReLU(0.2),
            nn.Conv1d(channels, channels, kernel_size, padding=dilation, dilation=dilation),
            nn.LeakyReLU(0.2),
            nn.Conv1d(channels, channels, kernel_size, padding=1)
        )

    def forward(self, x):
        return x + self.block(x)


class MelGANGenerator(nn.Module):
    def __init__(self, in_channels=80):
        super().__init__()
        self.pre_conv = nn.Conv1d(in_channels, 512, kernel_size=7, padding=3)

        self.upsample_layers = nn.ModuleList([
            nn.Sequential(
                nn.LeakyReLU(0.2),
                nn.ConvTranspose1d(512 // (2 ** i), 512 // (2 ** (i + 1)), 16, stride=8, padding=4)
            ) for i in range(3)
        ])

        self.res_blocks = nn.ModuleList([
            ResBlock(64, kernel_size=3, dilation=1) for _ in range(4)
        ])

        self.post_conv = nn.Sequential(
            nn.LeakyReLU(0.2),
            nn.Conv1d(64, 1, kernel_size=7, padding=3),
            nn.Tanh()
        )

    def forward(self, x):
        x = self.pre_conv(x)
        for layer in self.upsample_layers:
            x = layer(x)
        for block in self.res_blocks:
            x = block(x)
        return self.post_conv(x)

class MelGANDiscriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Conv1d(1, 16, 15, padding=7),
            nn.LeakyReLU(0.2),
            nn.Conv1d(16, 64, 41, stride=4, padding=20, groups=4),
            nn.LeakyReLU(0.2),
            nn.Conv1d(64, 256, 41, stride=4, padding=20, groups=16),
            nn.LeakyReLU(0.2),
            nn.Conv1d(256, 1024, 41, stride=4, padding=20, groups=64),
            nn.LeakyReLU(0.2),
            nn.Conv1d(1024, 1, kernel_size=5, padding=2)
        )

    def forward(self, x):
        return self.model(x)

from tqdm.notebook import tqdm

class Mel2WaveGANTrainer:
    def __init__(self, generator, discriminator, dataloader, device="cuda", max_batches_per_epoch=1500):
        self.gen = generator.to(device)
        self.dis = discriminator.to(device)
        self.data = dataloader
        self.device = device
        self.max_batches = max_batches_per_epoch

        self.opt_g = torch.optim.Adam(self.gen.parameters(), lr=1e-4)
        self.opt_d = torch.optim.Adam(self.dis.parameters(), lr=1e-4)

        self.l1_loss = nn.L1Loss()
        self.bce = nn.BCEWithLogitsLoss()
        self.stft_loss = STFTLoss()

    def collate_fn(self, batch):
        mels, wavs = zip(*batch)
        mel_tensor = torch.stack(mels)
        wav_tensor = torch.stack(wavs)
        return mel_tensor.to(self.device), wav_tensor.to(self.device)

    def train_epoch(self):
        self.gen.train()
        self.dis.train()
        pbar = tqdm(enumerate(self.data), desc="Training", leave=False)

        torch.cuda.empty_cache()

        for i, (mel, wav) in pbar:
            if i >= self.max_batches:
                break

            mel = mel.to(self.device)
            wav = wav.to(self.device)

            fake_wav = self.gen(mel)
            min_len = min(fake_wav.shape[-1], wav.shape[-1])
            fake_wav = fake_wav[:, :, :min_len]
            wav = wav[:, :, :min_len]

            real_pred = self.dis(wav)
            fake_pred = self.dis(fake_wav.detach())

            d_loss = self.bce(real_pred, torch.ones_like(real_pred)) + \
                     self.bce(fake_pred, torch.zeros_like(fake_pred))

            self.opt_d.zero_grad()
            d_loss.backward()
            self.opt_d.step()

            fake_pred = self.dis(fake_wav)
            adv_loss = self.bce(fake_pred, torch.ones_like(fake_pred))
            rec_loss = self.l1_loss(fake_wav, wav)
            stft_loss = self.stft_loss(fake_wav, wav)
            g_loss = adv_loss + rec_loss + stft_loss

            self.opt_g.zero_grad()
            g_loss.backward()
            self.opt_g.step()

            pbar.set_postfix({"G_loss": g_loss.item(), "D_loss": d_loss.item()})

        torch.cuda.empty_cache()
        gc.collect()
```

```python
from torch.utils.data import Dataset
import os
import torch
import torchaudio
from torchaudio.transforms import Resample
import random

class MelDataset(Dataset):
    def __init__(self, data_dir, t2s_converter, sample_rate=22050, segment_len_sec=2.0):
        self.wav_files = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith(".wav")]
        self.t2s = t2s_converter
        self.sample_rate = sample_rate
        self.segment_len = int(segment_len_sec * sample_rate)
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    def __len__(self):
        return len(self.wav_files)

    def __getitem__(self, idx):
        wav_path = self.wav_files[idx]
        txt_path = wav_path.replace(".wav", ".txt")

        waveform, sr = torchaudio.load(wav_path)
        if sr != self.sample_rate:
            waveform = Resample(orig_freq=sr, new_freq=self.sample_rate)(waveform)
        waveform = waveform.mean(0)
        waveform = waveform / waveform.abs().max()

        if waveform.shape[0] > self.segment_len:
            start = random.randint(0, waveform.shape[0] - self.segment_len)
            waveform = waveform[start:start + self.segment_len]
        else:
            waveform = torch.nn.functional.pad(waveform, (0, self.segment_len - waveform.shape[0]))

        with open(txt_path, "r") as f:
            text = f.read().strip()

        mel = self.t2s.text2spec(text)
        mel_tensor = torch.tensor(mel, dtype=torch.float32)

        if mel_tensor.shape[0] != 80:
            mel_tensor = mel_tensor.T

        T_max = 173
        T = mel_tensor.shape[1]
        if T > T_max:
            mel_tensor = mel_tensor[:, :T_max]
        elif T < T_max:
            mel_tensor = torch.nn.functional.pad(mel_tensor, (0, T_max - T))

        return mel_tensor.to(self.device), waveform.to(self.device).unsqueeze(0)
```

```python
t2s = TextToSpecConverter()
dataset = MelDataset("/content/LJSpeech-1.1/wavs", t2s)

gen = MelGANGenerator()
disc = MelGANDiscriminator()
trainer = Mel2WaveGANTrainer(gen, disc, None)

loader = DataLoader(dataset, batch_size=4, shuffle=True, drop_last=True, collate_fn=trainer.collate_fn)
trainer.data = loader  # обязательно присвоить
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

t2s.tts_handler.synthesizer.tts_model = t2s.tts_handler.synthesizer.tts_model.to(device)
t2s.tts_handler.device = device

gen = gen.to(device)
disc = disc.to(device)

dataset = MelDataset("/content/LJSpeech-1.1/wavs", t2s)
loader = DataLoader(dataset, batch_size=8, shuffle=True)

trainer = Mel2WaveGANTrainer(gen, disc, loader, device=device, max_batches_per_epoch=1200)
```

```python
for epoch in range(12):
    print(f"Epoch {epoch + 1}")
    trainer.train_epoch()

torch.save(gen.state_dict(), "melgan_generator.pth")
```
